```python
channels = []
for vid in youtube_data.video_ids:
    try:
        channels.append(youtube.videos().list(id= vid,
                       part='snippet').execute().get('items',[])[0]['snippet']['channelTitle'])
    except:
        channels.append(None)
```

```python
len(channels)
```

```python
youtube_data['channels'] = channels
```
